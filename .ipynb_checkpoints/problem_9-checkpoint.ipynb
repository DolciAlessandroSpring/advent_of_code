--- Day 9: Rope Bridge ---


This rope bridge creaks as you walk along it. You aren't sure how old it is, or whether it can even support your weight.

It seems to support the Elves just fine, though. The bridge spans a gorge which was carved out by the massive river far below you.

You step carefully; as you do, the ropes stretch and twist. You decide to distract yourself by modeling rope physics; maybe you can even figure out where not to step.

Consider a rope with a knot at each end; these knots mark the head and the tail of the rope. If the head moves far enough away from the tail, the tail is pulled toward the head.

Due to nebulous reasoning involving Planck lengths, you should be able to model the positions of the knots on a two-dimensional grid. Then, by following a hypothetical series of motions (your puzzle input) for the head, you can determine how the tail will move.

Due to the aforementioned Planck lengths, the rope must be quite short; in fact, the head (H) and tail (T) must always be touching (diagonally adjacent and even overlapping both count as touching):

````python
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)

txt = open("/Users/adolci/Documents/Python/private/advent_of_code/problem_9_input.txt", 'r').read()
txt = txt.split("\n")[:-1]
````

````python
steps = list(''.join(
    [a[0] * int(a[2]) for a in txt]
))
````

````python
def move_H(h_pos, step):
    # H can move by one step N, S, E or W.
    if step == 'D':
        return [sum(x) for x in zip(h_pos, [1, 0])]
    elif step == 'U':
        return [sum(x) for x in zip(h_pos, [-1, 0])]
    elif step == 'L':
        return [sum(x) for x in zip(h_pos, [0, -1])]
    else:
        return [sum(x) for x in zip(h_pos, [0, 1])]
    
def update_board_after_H(board_move, h_new_pos, h_pos):
    # H has two options: either it's moving to an empty place
    # or it's moving on top of T.
    
    if board_move[h_new_pos[0], h_new_pos[1]] == 0:
        board_move[h_new_pos[0], h_new_pos[1]] = 2
    elif board_move[h_new_pos[0], h_new_pos[1]] == 1:
        board_move[h_new_pos[0], h_new_pos[1]] = 21
    else:
        raise ValueError('Unexpected move of H.')
    
    # put a 0 in the old position, except if you have a 21
    if board_move[h_pos[0], h_pos[1]] == 21:
        board_move[h_pos[0], h_pos[1]] = 1
    elif board_move[h_pos[0], h_pos[1]] == 2:
        board_move[h_pos[0], h_pos[1]] = 0
    else:
        raise ValueError('Unexpected move of H xxx.')
    return board_move

def move_T(t_pos, h_new_pos, h_pos):

    # first decide whether T should move or not
    dist_H_T = np.linalg.norm(np.array(h_new_pos) - np.array(t_pos))

    if dist_H_T >= 2:
        # if T moves, it goes where H was
        t_new_pos = h_pos
    else:
        t_new_pos = t_pos
    return t_new_pos

def update_board_after_T(board_move, board_visits, t_new_pos, t_pos):
    
    # move T from 0 --> to 1. Note that if T didn't move, the code is not doing anything in the end

    if t_new_pos != t_pos:
        board_move[t_pos[0], t_pos[1]] = 0
        board_move[t_new_pos[0], t_new_pos[1]] = 1
        board_visits[t_new_pos[0], t_new_pos[1]] = 1

    return board_move, board_visits
````

##### BOARD MOVE

0: empty position

2: Head

1: Tail

21: Head and Tail on the same position


##### BOARD VISITS

0: not visited yet

1: visited

````python
N = 500
half_N = int(N/2)

board_move = np.zeros((N, N))
board_visits = np.zeros((N, N))

board_move[half_N][half_N] = 21
board_visits[half_N][half_N] = 1
h_pos = [half_N, half_N]
t_pos = [half_N, half_N]
````

````python
for step in steps:

#     print(step)

    h_new_pos = move_H(h_pos, step)
    board_move = update_board_after_H(board_move, h_new_pos, h_pos)
#     print("Move H:")
#     print(board_move)
#     print("")
    
    t_new_pos = move_T(t_pos, h_new_pos, h_pos)
    board_move, board_visits = update_board_after_T(board_move, board_visits, t_new_pos, t_pos)
#     print("Move T:")
#     print(board_move)
#     print("")
#     print(board_visits)
#     print("")
    
    h_pos = h_new_pos
    t_pos = t_new_pos
#     print("end of epoch")
#     print("")
````
